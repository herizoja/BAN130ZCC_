### We have to import the datetime module to work with dates

```python
#Import datetime module
import datetime
x = datetime.datetime.now()
print(x)
```

The date contains year, month, day, hour, minute, second, and microsecond.

### Common formats are:
    
1. %Y: 4-digit year (2021)
2. %y: 2-digit year (21)
3. %m: 2-digit month (01)
4. %d: 2-digit day of the month (18)
5. %A: weekday (Monday)
6. %a: abbreviated weekday (Mon)
7. %B: month (January)
8. %b: abbreviated month (Jan)

```python
#import datetime

x = datetime.datetime(2022, 6, 1)

print(x.strftime("%B"))
print(x.strftime("%A"))
```

Once we have dates stored in dataset, we can extract the weekday, month, or quarter with the functions weekdays(), months(), quarters(). To extract the numeric values for days, months, and years, we can use the format() function, noting that the output is as text. For example,

```python
# let us import the flights_subset dataset.
import pandas as pd
df = pd.read_csv("flights_subset.csv")
df.head()
```

```python
df.info()
```

```python
# As we can see the date colmn is of type integer and we have to convert it to datetime format.
#We can also convert and change the format
df['date']= pd.to_datetime(df['date'], format='%d%m%Y')
df.info()

# we can also use astype function
#df['date'] = df['date'].astype('datetime64[ns]')
#df.info()


df.head()
```

```python
# Now we can work with this column and do some extraction. We can extract the year
df['date'].dt.year
# Extract the month
df['date'].dt.month
df['date'].dt.day
# WE CAN ALSO ACCESS THE DAY NAME 
df['date'].dt.day_name()
```
